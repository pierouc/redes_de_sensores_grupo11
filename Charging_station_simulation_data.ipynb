```python
import pandas as pd

data = pd.read_csv('ChargePoint_Data_CY20Q4.csv',low_memory=False)

data['Start Date'] = pd.DataFrame([d.split(' ')[0] for d in data['Start Date']])

most_repeated = data['Start Date'].value_counts(sort='ascendant').index[0]

day_most_used = data[data['Start Date'] == most_repeated]

day_most_used = day_most_used.reset_index()

clean_data = list(day_most_used['Energy (kWh)'])

clean_data
```

```python
t_transport = "websockets"
t_port = 80
```

```python
# Create the topic string.
topic = "channels/" + channel_ID + "/publish"
```

```python
for d in clean_data:
    # build the payload string.
    payload = "field2=" + str(d)
    time.sleep(1.5)
    # attempt to publish this data to the topic.
    try:
        print ("Writing Payload = ", payload," to host: ", mqtt_host, " clientID= ", mqtt_client_ID, " User ", mqtt_username, " PWD ", mqtt_password)
        publish.single(topic, payload, hostname=mqtt_host, transport=t_transport, port=t_port, client_id=mqtt_client_ID, auth={'username':mqtt_username,'password':mqtt_password})
    except (keyboardInterrupt):
        break
    except Exception as e:
        print (e) 
```
